```python
import string
punctuation = string.punctuation
punctuation
```

```python
# Stemming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#test code 
ps.stem("beautiful")
```

```python
def text_transformer(text):
    
    # convert all text in lower case
    text = text.lower()
    
    #Tokenization
    y=[]
    text = nltk.word_tokenize(text)
    
    for i in text:
        if i.isalnum():
            y.append(i)
    text = y[:]
    y.clear()
    
    # Removing stopwords and punctuations
    for i in text:
        if (i not in stopwords) and (i not in punctuation):
            y.append(i)
    text=y[:]
    y.clear()
    
    
    # apply stemming
    
    for i in text:
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
import ast

def convert_cast(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i["name"])
        
            counter+=1
        else:
            break
    return l
(movies.cast.apply(convert_cast)[0])
```

```python
import ast  # list ko dictionary me convert krega


def convert_crew(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i["name"])
        
            counter+=1
        else:
            break
    return l
(movies.cast.apply(convert_cast)[0])
```
