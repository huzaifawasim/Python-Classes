```python
def func(arr,task=""):
    if task=="sum":
        return sum(arr)
    elif task=="mult":
        res =1
        for num in arr:
            res*=num
        return res 
    elif task=="max":
        return max(arr)
    
    elif task=="min":
        return min(arr)
    elif task=="even":
        even=[]
        for num in arr:
            if num %2==0 :
                even.append(num)
        return even
    elif task=="odd":
        odd=[]
        for num in arr:
            if num %2!=0 :
                odd.append(num)
        return odd
    
    
            
```
